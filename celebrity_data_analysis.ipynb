```python
import pandas as dp
df= pd.read_csv(
    "/kaggle/input/9000-popular-people-in-cinema/popular_people.csv"
)
df = df.drop_duplicates()
df = df.dropna()
print(df.info())
print(df.describe())
# phân tích dữ liệu
print(
    df["known_for_department"]
    .value_counts()
)
# vẽ chart
import matplotlib.pyplot as plt

df["known_for_department"] \
.value_counts() \
.head(10) \
.sort_values() \
.plot(kind="barh")
plt.title("department distribution")
plt.xlabel("count")
plt.ylabel("department")
plt.show()
# bar chart
df['gender'] \
.value_counts() \
.head(10) \
.sort_values() \
.plot(kind="barh")
plt.title("department gender")
plt.xlabel("count")
plt.ylabel("Department")
plt.show()
# line chart
top_popularity = (
    df["popularity"]
    .sort_values(ascending=False)
    .head(100)
)

plt.plot(top_popularity.values)

plt.title("Top Popularity Trend")

plt.show()
# top người nổi tiếng nhất
top_people =df.sort_values(
    by="popularity",
    ascending=False

)
print(
    top_people[["name","popularity"]]
    .head(10)
)
# vẽ histogram populatity
df["popularity"].plot(kind="hist",bins = 30)
plt.title("historygogram")
plt.show()
# đếm giới tính
print(df["gender"]
      .value_counts()
      )
#pie chart
df['gender']\
.value_counts()\
.plot(
    kind="pie",
    autopct ="%1.1f%%"
)
plt.ylabel("")
plt.show()
```
